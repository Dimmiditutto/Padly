```python
import subprocess
import sys

tests = [
    'tests/test_booking_api.py::test_public_cancellation_outside_refund_window_triggers_automatic_refund',
    'tests/test_booking_api.py::test_public_cancellation_within_24h_does_not_refund_paid_booking',
    'tests/test_payment_flows.py::test_public_cancellation_persists_failed_refund_without_false_success',
    'tests/test_payment_flows.py::test_booking_confirmation_email_mentions_refund_only_before_cutoff',
    'tests/test_payment_flows.py::test_public_cancellation_email_mentions_no_refund_inside_cutoff',
]

for test in tests:
    cmd = ['../.venv/bin/python', '-m', 'pytest', test, '-q']
    result = subprocess.run(cmd, cwd='/workspaces/PadelBooking/backend', text=True, capture_output=True)
    print('TEST=', test)
    print('EXIT_CODE=', result.returncode)
    if result.returncode == 0:
        print(result.stdout)
        continue
    if result.stdout:
        print(result.stdout[-4000:])
    if result.stderr:
        print(result.stderr[-4000:], file=sys.stderr)
    break
```

```python
import subprocess
import sys

commands = [
    (['../.venv/bin/python', '-m', 'pytest', 'tests', '-q'], '/workspaces/PadelBooking/backend'),
    (['npm', 'run', 'build'], '/workspaces/PadelBooking/frontend'),
    (['npm', 'run', 'test:run'], '/workspaces/PadelBooking/frontend'),
]

for cmd, cwd in commands:
    result = subprocess.run(cmd, cwd=cwd, text=True, capture_output=True)
    print('CMD=', ' '.join(cmd))
    print('CWD=', cwd)
    print('EXIT_CODE=', result.returncode)
    combined = result.stdout
    if result.stderr:
        combined = combined + ('\n' if combined else '') + result.stderr
    if result.returncode == 0:
        print(combined[-4000:] if len(combined) > 4000 else combined)
        continue
    print(combined[-6000:] if len(combined) > 6000 else combined)
    raise SystemExit(result.returncode)
```
